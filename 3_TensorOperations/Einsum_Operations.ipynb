TensorFlow's tf.einsum function is a powerful tool for tensor operations using Einstein Summation Notation. It enables:

✅ Efficient Computation

✅ Concise & Readable Expressions

✅ Broad Applications in Deep Learning, Physics, and Data Science



### Tensor Definitions

```python
# Define tensors for operations
A = tf.constant([[1, 2], [3, 4]], dtype=tf.float32)  # 2x2 Matrix
B = tf.constant([[5, 6], [7, 8]], dtype=tf.float32)  # 2x2 Matrix
v = tf.constant([1, 2], dtype=tf.float32)  # Vector
M = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)  # 2x3 Matrix
N = tf.constant([[7, 8], [9, 10], [11, 12]], dtype=tf.float32)  # 3x2 Matrix
```

## Einsum Operations

# Permutation of Tensors

* Rearrange tensor axes using einsum.
* "ij->ji" swaps rows and columns.



```python
# Transpose a Matrix
transpose_A = tf.einsum("ij->ji", A)
print("Transposed Matrix:\n", transpose_A.numpy())
```

# Summation

* Sum of all elements
* "ij->" collapses all dimensions.

```python
sum_A = tf.einsum("ij->", A)  # Sum of all elements
print("Sum of all elements:", sum_A.numpy())
```

# Column Sum

* "ij->j" sums across rows.

```python
col_sum = tf.einsum("ij->j", A)
print("Column-wise Sum:", col_sum.numpy())
```

# Row Sum

* "ij->i" sums across columns.

```python
row_sum = tf.einsum("ij->i", A)
print("Row-wise Sum:", row_sum.numpy())
```

# Matrix-Vector Multiplication

* "ij,j->i" multiplies each row of A with vector v.

```python
mat_vec_mul = tf.einsum("ij,j->i", A, v)
print("Matrix-Vector Multiplication:\n", mat_vec_mul.numpy())
```

# Matrix-Matrix Multiplication

* "ik,kj->ij" performs standard matrix multiplication.

```python
mat_mat_mul = tf.einsum("ik,kj->ij", M, N)
print("Matrix-Matrix Multiplication:\n", mat_mat_mul.numpy())
```

# Dot Product (First row with First row of Matrix)

* "i,i->" computes the dot product.

```python
dot_product_row = tf.einsum("i,i->", A[0], A[0])
print("Dot Product (First row with First row):", dot_product_row.numpy())
```

# Dot Product with Matrix

* "ij,ij->" computes the dot product across all elements.

```python
dot_product_matrix = tf.einsum("ij,ij->", A, B)
print("Dot Product with Matrix:", dot_product_matrix.numpy())
```

# Hadamard Product (Element-wise Multiplication)

* "ij,ij->ij" performs element-wise multiplication.

```python
hadamard = tf.einsum("ij,ij->ij", A, B)
print("Hadamard Product:\n", hadamard.numpy())
```

# Outer Product

* "i,j->ij" computes the outer product.

```python
outer_product = tf.einsum("i,j->ij", v, v)
print("Outer Product:\n", outer_product.numpy())
```

# Batch Matrix Multiplication

* "bij,bjk->bik" performs batch-wise matrix multiplication.

```python
batch_A = tf.random.normal((2, 3, 4))
batch_B = tf.random.normal((2, 4, 5))

batch_matmul = tf.einsum("bij,bjk->bik", batch_A, batch_B)
print("Batch Matrix Multiplication Shape:", batch_matmul.shape)
```

# Bilinear Transformation

* "ijk,ik->ij" applies a bilinear transformation.

```python
W = tf.random.normal((3, 3, 3))
X = tf.random.normal((3, 3))

bilinear = tf.einsum("ijk,ik->ij", W, X)
print("Bilinear Transformation Shape:", bilinear.shape)
```

# Weighted Sum

* "i,ij->j" computes weighted sum across rows.

```python
weights = tf.constant([0.1, 0.2, 0.3], dtype=tf.float32)
values = tf.constant([[1, 2, 3], [4, 5, 6], [7, 8, 9]], dtype=tf.float32)

weighted_sum = tf.einsum("i,ij->j", weights, values)
print("Weighted Sum:\n", weighted_sum.numpy())
```

# Matrix Diagonal

* "ii->i" selects diagonal elements.

```python
matrix_diag = tf.einsum("ii->i", A)
print("Matrix Diagonal:", matrix_diag.numpy())
```

# Matrix Trace

* "ii->" sums diagonal elements.

```python
matrix_trace = tf.einsum("ii->", A)
print("Matrix Trace:", matrix_trace.numpy())
```
